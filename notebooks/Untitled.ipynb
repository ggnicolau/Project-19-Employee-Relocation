```python
df_full = df_full[['Potencial Bruto', 'Social', 'Motivacional', 'Raciocínio', 'Cultura pontuação', 'label']].dropna()

x = df_full[['Potencial Bruto', 'Social', 'Motivacional', 'Raciocínio', 'Cultura pontuação']] # todas as colunas ficou pior
x = x[['Raciocínio', 'Cultura pontuação']]

y = df_full['label']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=0.1)
```

```python
#Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
#Create KNN Classifiers
knn1 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=1)
knn5 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=5)
knn10 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=10)
knn15 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=15)
knn20 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=20)
knn25 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=25)
knn30 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=30)
knn35 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=35)
knn40 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=40)
knn45 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=45)
knn50 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=50)
knn70 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=70)
knn500 = KNeighborsClassifier(metric = 'sqeuclidean', weights = 'uniform', n_neighbors=500)

#Train the model using the training sets
knn1.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn1.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("\n\nK=1, Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")
#Train the model using the training sets
knn5.fit(x_train, y=y_train)
#Predict the response for test dataset
y_pred = knn5.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=5 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")
#Train the model using the training sets
knn10.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn10.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=10 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")
#Train the model using the training sets
knn15.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn15.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=15 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")
#Train the model using the training sets
knn20.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn20.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=20 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")
#Train the model using the training sets
knn25.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn25.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=25 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")
#Train the model using the training sets
knn30.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn30.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=30 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")
#Train the model using the training sets
knn35.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn35.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=35 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")

#Train the model using the training sets
knn40.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn40.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=40 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")

#Train the model using the training sets
knn45.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn45.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=45 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")

#Train the model using the training sets
knn50.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn50.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=50 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")

#Train the model using the training sets
knn70.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn70.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=70 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")


#Train the model using the training sets
knn500.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = knn500.predict(x_test)
# Model Accuracy, how often is the classifier correct?
print("K=500 Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,1), "%")
```
